열 요소
* 각 행은 한 건물이 아니라 한 지역의 위치, 그 지역의 평균 건물 나이,
인구 수 등을 의미.

 0   longitude : 지역 경도 (x, 가로)  

 1   latitude : 지역 위도 (y, 세로)

 2   housing_median_age : 건물 지어진 나이 (숫자가 적을 수록 최신 건물)

 3   total_rooms : 한 블록 내 총 방 수

 4   total_bedrooms :한 블록 내 총 침실 개수

 5   population : 한 블록 내 인구 수

 6   households : 한 블록 내 거주 세대(가구수)

 7   median_income : 중위 소득

 8   median_house_value : 중위 주택 가격

 9   ocean_proximity : 바다와 가까운지. 만이나 해안선 근처, 섬, 내륙 쪽인지 알 수 있음.

 => 'NEAR BAY', '<1H OCEAN', 'INLAND', 'NEAR OCEAN', 'ISLAND'

```python
#위도 경도를 x, y축으로 하여 어느 쪽에 건물이 몰려있는지 확인
import pandas as pd
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 7))
plt.scatter(df['longitude'], df['latitude'], alpha=0.2, s=10)
```

```python
#건물이 몰려있는 원인 찾기 -> 예측1: 해변 근처에 건물이 많을 것...
#                          -> 결과: 해변 근처에 건물이 많음. but 내륙에도 몰려있는 곳 존재.
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 7))
plt.scatter(df['longitude'], df['latitude'], color='lightgray', alpha=0.2, label='Other')

# ocean proximity
highlight_categories = ['<1H OCEAN', 'NEAR BAY', 'NEAR OCEAN']

colors = {
    '<1H OCEAN': 'blue',
    'NEAR BAY': 'green',
    'NEAR OCEAN': 'purple'
}

# 조건에 맞는 데이터만 점 색깔 다르게 하기
for category in highlight_categories:
    subset = df[df['ocean_proximity'] == category]
    plt.scatter(subset['longitude'], subset['latitude'],
                color=colors[category], alpha=0.3, label=category)

plt.show()
```

```python
#건물이 몰려있는 원인 찾기 -> 예측2: 내륙 쪽은 유동인구수가 많아서...?
```
